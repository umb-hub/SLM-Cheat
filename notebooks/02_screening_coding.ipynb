# 02 — Screening & Coding Overview

Distribution of all four coding fields, cross-tabulations, and a Sankey flow from temporal phase through contribution type to study approach.

**Requires:** `data/processed/corpus_clean.parquet` (run NB00 first)

```python
# ── CONFIG ────────────────────────────────────────────────────────────────────
PROCESSED_DIR = "../data/processed/"
OUTPUTS_DIR   = "../outputs/"
```

```python
# ── LOAD ──────────────────────────────────────────────────────────────────────
import sys
sys.path.insert(0, "..")

import pandas as pd
import plotly.graph_objects as go
from IPython.display import display

from scripts.utils import (
    load_processed, freq_table, crosstab_with_pct,
    save_figure, STYLE,
)

df = load_processed("corpus_clean")
N  = len(df)
print(f"Corpus: N={N}")
```

## 1 · Relevance Level

```python
rl_freq = freq_table(df["relevance_level"])
print(f"Relevance Level distribution (N={N}):")
display(rl_freq)

rl_colors = [STYLE["relevance_level"].get(v, "#9E9E9E") for v in rl_freq["value"]]

fig = go.Figure(go.Pie(
    labels=[f"{v}\nN={n} ({p}%)" for v, n, p in zip(rl_freq.value, rl_freq.N, rl_freq.pct)],
    values=rl_freq["N"],
    hole=0.4,
    marker_colors=rl_colors,
    textinfo="label",
    textfont_size=12,
))
fig.update_layout(
    title=f"Relevance Level Distribution (N={N})",
    height=450,
    showlegend=False,
)
fig.show()
save_figure(fig, "02_relevance_donut")
```

```python
ct_freq = freq_table(df["contribution_type"])
print(f"Contribution Type distribution (N={N}):")
display(ct_freq)

ct_colors = [STYLE["contribution_type"].get(v, "#9E9E9E") for v in ct_freq["value"]]

fig = go.Figure(go.Pie(
    labels=[f"{v}\nN={n} ({p}%)" for v, n, p in zip(ct_freq.value, ct_freq.N, ct_freq.pct)],
    values=ct_freq["N"],
    hole=0.4,
    marker_colors=ct_colors,
    textinfo="label",
    textfont_size=11,
))
fig.update_layout(
    title=f"Contribution Type Distribution (N={N})",
    height=480,
    showlegend=False,
)
fig.show()
save_figure(fig, "02_contribution_donut")
```

## 4 · Cross-Tabulations

```python
print("── Relevance Level × Contribution Type ──")
raw_rl_ct, pct_rl_ct = crosstab_with_pct(df, "relevance_level", "contribution_type")
print("Count:")
display(raw_rl_ct)
print("Row % (row = relevance level):")
display(pct_rl_ct)

print("\n── Contribution Type × Study Approach ──")
raw_ct_sa, pct_ct_sa = crosstab_with_pct(df, "contribution_type", "study_approach")
print("Count:")
display(raw_ct_sa)
print("Row % (row = contribution type):")
display(pct_ct_sa)

# Save freq tables
import pandas as pd
coding_freq = pd.concat([
    freq_table(df["relevance_level"]).assign(field="relevance_level"),
    freq_table(df["contribution_type"]).assign(field="contribution_type"),
    freq_table(df["study_approach"]).assign(field="study_approach"),
], ignore_index=True)
coding_freq.to_parquet(PROCESSED_DIR + "coding_freq_tables.parquet", index=False)
print("\n✓ coding_freq_tables.parquet exported")
```

## 5 · Coding Flow Sankey: Phase → Contribution Type → Study Approach

```python
# Build node list: phases (left), contribution types (middle), study approaches (right)
phases  = sorted(
    [p for p in df["phase"].unique() if str(p).startswith("Phase")],
    key=lambda p: int(p.split()[1]),
)
ct_vals = sorted(df["contribution_type"].dropna().unique(), key=lambda x: df["contribution_type"].value_counts().get(x, 0), reverse=True)
sa_vals = sorted(df["study_approach"].dropna().unique(), key=lambda x: df["study_approach"].value_counts().get(x, 0), reverse=True)

all_nodes = phases + ct_vals + sa_vals
node_idx  = {n: i for i, n in enumerate(all_nodes)}

phase_counts = df["phase"].value_counts().to_dict()
ct_counts    = df["contribution_type"].value_counts().to_dict()
sa_counts    = df["study_approach"].value_counts().to_dict()

node_labels = (
    [f"{p}<br>(N={phase_counts.get(p, 0)})" for p in phases]
    + [f"{c}<br>(N={ct_counts.get(c, 0)})" for c in ct_vals]
    + [f"{s}<br>(N={sa_counts.get(s, 0)})" for s in sa_vals]
)
node_colors = (
    [STYLE["phase"].get(p, "#90CAF9") for p in phases]
    + [STYLE["contribution_type"].get(c, "#BDBDBD") for c in ct_vals]
    + [STYLE["study_approach"].get(s, "#BDBDBD") for s in sa_vals]
)

# x/y hints for arrangement="snap":
# x places nodes in the correct column; y sets the vertical order within it.
# Phase nodes are pinned to x=0.001 and spaced evenly top→bottom.
def _spread(n, lo=0.05, hi=0.95):
    return [0.5] if n <= 1 else [round(lo + i * (hi - lo) / (n - 1), 4) for i in range(n)]

n_ph, n_ct, n_sa = len(phases), len(ct_vals), len(sa_vals)
node_x = [0.001] * n_ph + [0.45]  * n_ct + [0.99]  * n_sa
node_y = _spread(n_ph)  + _spread(n_ct)   + _spread(n_sa)

df_valid = df[
    df["phase"].isin(phases)
    & df["contribution_type"].notna()
    & df["study_approach"].notna()
].copy()

excluded = len(df) - len(df_valid)
if excluded > 0:
    print(f"Excluded from Sankey: {excluded} rows (missing phase / coding fields)")

link_src, link_tgt, link_val, link_lbl = [], [], [], []

for (ph, ct), grp in df_valid.groupby(["phase", "contribution_type"], sort=False):
    if ph in node_idx and ct in node_idx:
        cnt = len(grp)
        link_src.append(node_idx[ph]); link_tgt.append(node_idx[ct])
        link_val.append(cnt);          link_lbl.append(str(cnt))

for (ct, sa), grp in df_valid.groupby(["contribution_type", "study_approach"], sort=False):
    if ct in node_idx and sa in node_idx:
        cnt = len(grp)
        link_src.append(node_idx[ct]); link_tgt.append(node_idx[sa])
        link_val.append(cnt);          link_lbl.append(str(cnt))

fig = go.Figure(go.Sankey(
    arrangement="snap",
    node=dict(
        label=node_labels,
        color=node_colors,
        x=node_x,
        y=node_y,
        pad=15,
        thickness=18,
        line=dict(color="#555", width=0.5),
    ),
    link=dict(
        source=link_src,
        target=link_tgt,
        value=link_val,
        label=link_lbl,
        hovertemplate="%{source.label} → %{target.label}<br>N=%{value}<extra></extra>",
    ),
))
fig.update_layout(
    title_text=f"Coding Flow: Phase → Contribution Type → Study Approach (N={len(df_valid)})",
    font_size=12,
    height=620,
)
fig.show()
save_figure(fig, "02_coding_sankey")
```
